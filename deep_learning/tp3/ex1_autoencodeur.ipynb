# Dataset

```python
import numpy as np
```

```python
X_train = np.random.binomial(1, 0.5, (5000, 20))
X_val = np.random.binomial(1, 0.5, (1000, 20))
X_test = np.random.binomial(1, 0.5, (1000, 20))
```

# Model

```python
from keras import Sequential, Input
from keras.layers import Dense
from keras import activations, losses, metrics
from keras.optimizers import Adam
```

```python
model = Sequential()
model.add(Input((20, )))
model.add(Dense(15, activation = activations.relu))
model.add(Dense(10, activation = activations.relu))
model.add(Dense(15, activation = activations.relu))
model.add(Dense(20, activation = activations.sigmoid))
model.compile(optimizer = Adam(), loss = losses.binary_crossentropy, metrics = [metrics.binary_crossentropy])
history = model.fit(x = X_train, y = X_train, validation_data = (X_val, X_val), batch_size = 256, epochs = 800,
                    verbose = 0)
```

```python
import matplotlib.pyplot as plt

plt.plot(range(len(history.history["binary_crossentropy"])), history.history["binary_crossentropy"])
plt.plot(range(len(history.history["binary_crossentropy"])), history.history["val_binary_crossentropy"])
```
